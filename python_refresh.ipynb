# התניות
התניות הן משפטים לוגים מסוג


```
if condition:
  somthing to do
  elif condition:
  anotherthing to do
  else:
  some anotherthing to do
```

נבדוק מה האם תוצאות לוגיות של אמת זהות לאחת ותוצאות שקריות זהות לאפס




````python
if 1: # in this code the condition was replaced by 1
  print("1 is true")
else:
  print("1 is not true")
````

האם אפס הוא שקר?

````python
if not 0: # in this code the condition was replaced by 0
  print("0 is false")
else:
  print("0 is true")
````

````python
if 2:
  print("also 2 is true")
````

# פונקציות
פונקציה מוגדרת בצורה הבאה:


```
def my_function(variables,....,variable=default_value):
  code
  return value_to_return
```

כדי לעבוד בצורה יעילה ונקיה בפיתון, נרצה לפרק כל משימה למשימות קטנות ופשוטות. ניצור פונקציה לכל משימה פשוטה, ואז נקרא לפונקציות מתוך הקוד הראשי. לצורך התרגיל, נכתוב פונקציה שמחברת שני מספרים. במידה ויקראו לפונקציה רק עם משתנה אחד, היא תוסיף לו את הערך אפס, כלומר תחזיר את אותו המספר

````python
def summing(x,y=0):
  return x+y
````

נסו את הפונקציה. שימו לב שכדי שהיא תהיה מוגדרת, חייבים להריץ את התא הקודם

````python
print(summing(1,2))
print(summing(3))
````
